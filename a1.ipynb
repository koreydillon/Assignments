The histogram plot allows us discover the underlying frequency distribution of tenure data associated with the data scientists including paid and unpaid members in the club.
```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_data2(data):
    """
    Plots a histogram of tenured years for data scientists in the club, distinguishing between paid and unpaid members.

    Args:
        data: A list of tuples, where each tuple represents a data scientist with the following elements:
              - 'tenured_years': Number of years the scientist has been tenured
              - 'salary': Annual salary of the scientist
              - 'paid_account': Boolean, 1 if the scientist has a paid account, 0 otherwise

    Returns:
        None
    """

    if not data:
        print("Error: No valid data points found.")
        return

    tenure_paid = [d[0] for d in data if d[2] == 1] 
    tenure_unpaid = [d[0] for d in data if d[2] == 0] 

    plt.figure(figsize=(10, 6))
    plt.hist(tenure_paid, bins=10, alpha=0.7, label='Paid Members')
    plt.hist(tenure_unpaid, bins=10, alpha=0.7, label='Unpaid Members')
    plt.xlabel('Tenured Years')
    plt.ylabel('Number of Data Scientists')
    plt.title('Distribution of Tenured Years')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_data2(data)
```

# Problem 3

For this problem, you need to write a function **plotData3** to make a figure with two subplots.  One is a bar plot, and the other is a histogram plot. The bar plot can conveniently support users to visualize how many accounts are paid and how many are unpaid. The specific numbers of accounts for each type should be presented. The histogram plot allows us discover the underlying frequency distribution of salary data associated with the two types of data scientists (paid and unpaid).  Your plot should be similar to the figure that can be displayed below by running the code cell.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_data3(data):
    """
    Creates a figure with two subplots:
        1. Bar plot showing the number of paid and unpaid accounts.
        2. Histogram showing the salary distribution for paid and unpaid members.

    Args:
        data: A list of tuples, where each tuple represents a data scientist with the following elements:
              - 'tenured_years': Number of years the scientist has been tenured
              - 'salary': Annual salary of the scientist
              - 'paid_account': Boolean, 1 if the scientist has a paid account, 0 otherwise

    Returns:
        None
    """

    if not data:
        print("Error: No valid data points found.")
        return

    paid_accounts = [d for d in data if d[2] == 1]
    unpaid_accounts = [d for d in data if d[2] == 0]

   
    account_types = ['Paid', 'Unpaid']
    account_counts = [len(paid_accounts), len(unpaid_accounts)]

    
    salaries_paid = [d[1] for d in paid_accounts]
    salaries_unpaid = [d[1] for d in unpaid_accounts]


    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

 
    ax1.bar(account_types, account_counts)
    ax1.set_xlabel('Account Type')
    ax1.set_ylabel('Number of Accounts')
    ax1.set_title('Paid vs. Unpaid Accounts')
    for i, v in enumerate(account_counts):
        ax1.text(i, v + 0.5, str(v), ha='center')  


    ax2.hist(salaries_paid, bins=10, alpha=0.7, label='Paid Members')
    ax2.hist(salaries_unpaid, bins=10, alpha=0.7, label='Unpaid Members')
    ax2.set_xlabel('Salary')
    ax2.set_ylabel('Number of Data Scientists')
    ax2.set_title('Salary Distribution')
    ax2.legend()

    plt.show()
    
plot_data3(data)
```

# Problem 4 Community Data

For this problem, you need to write a function **barCommunityData** make a bar plot that shows numbers of counties in states.  Your plot should be similar to the figure that can be displayed below by running the code cell that has the Image object.  To approach this problem, you need to write the function definition that includes statements to parse the data from the provided file named "county.csv". For this assignment, you are **not** allowed to use pandas library to process the data file.   You are required to read the file row by row and extract the data to address the question.  In your plot, the specific numbers of counties in the states should be presented.

For the reference result plot, it shows the numbers of counties in Arizona, California, and Texas. However, your **barCommunityData** applications should allow any other states on the second argument.  

```python
import matplotlib.pyplot as plt

def bar_community_data(filename):
    """
    Creates a bar plot showing the number of counties in each state.

    Args:
        filename: The path to the 'county.csv' file.

    Returns:
        None
    """

    states = {}
    with open(filename, 'r') as file:
        next(file) 
        for line in file:
            line = line.strip()
            if line:
                state = line.split(',')[1]  
                if state in states:
                    states[state] += 1
                else:
                    states[state] = 1

    state_names = list(states.keys())
    county_counts = list(states.values())

    plt.figure(figsize=(12, 6))
    plt.bar(state_names, county_counts)
    plt.xlabel('State')
    plt.ylabel('Number of Counties')
    plt.title('Number of Counties per State')
    plt.xticks(rotation=45, ha='right')  

    
    for i, v in enumerate(county_counts):
        plt.text(i, v + 0.5, str(v), ha='center', fontsize=8)

    plt.tight_layout()
    plt.show()


bar_community_data('county.csv')
```
